```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model = "gpt-4.1-nano", temperature=0.2)
```

Q1: Clarity & Directness

```python
prompt_1 = "Define Artificial Intelligence in one sentence"
result = llm.invoke(prompt_1)
print(result.content) 
```

Q2: Specifying Output Format

```python
prompt_2 = "List 3 benefits of regular excercise using distinct bullet points"
result = llm.invoke(prompt_2)
print(result.content)
```

Q3: Setting a Persona/Role

```python
prompt_3 = "You are an expert economist. Explain GDP of a country"
result = llm.invoke(prompt_3)
print(result.content)
```

Q4: Providing Context

```python
context = "The Amazon rainforest is the largest rainforest in the world, renowned for its incredible biodiversity."
prompt_4 = f"Given the following: '{context}' \nWhat is the main characteristic mentioned about the Amazon rainforest?"
result = llm.invoke(prompt_4)
print(f"Prompt: {prompt_4}")
print(f"Response: {result.content}")
```

Q5: Setting Constraints (Length)

```python
prompt_5 = "Summarise the story of 'Cinderela' in exactly two sentence"
result = llm.invoke(prompt_5)
print(result.content)
```

Q6: Negative Constraints (What NOT to do)

```python
prompt_6 = "Describe a common dog breed, but do not mention its fur or color"
result = llm.invoke(prompt_6)
print(result.content)
```

Q7: Example/Few-shot Prompting

```python
prompt_7 = '''
Input:Apple
Output:Fruit

Input:Carrot
Output:Vegetable

Input:Mango
Output:
'''
result = llm.invoke(prompt_7)
print(result.content)
```

Q8: Asking for Explanation/Step-by-step reasoning

```python
prompt_8 = "Explain the process of Photosynthesis for a 10year old in step by step instructions"
results = llm.invoke(prompt_8)
print(results.content)
```

Q9: Iterative Prompting / Refining a Prompt

```python
# Write an initial prompt
initial_prompt_9 = "Write a short poem on nature"
initial_result = llm.invoke(initial_prompt_9)
print(initial_result.content)

# Write a refined prompt of initial prompt
refined_prompt_9 = "Write a short peom about a flowing water, focussing on its movement and sound"
refined_result = llm.invoke(refined_prompt_9)
print(refined_result.content)
```

Q10: Combining Technique

```python
prompt_10 = """
You are a friendly travel agent.
Your task is recommend 3 must visit places in Paris for the first time traveler.
Each recommendation must be short, not more than 15 words.
"""
result = llm.invoke(prompt_10)
print(result.content)
```
