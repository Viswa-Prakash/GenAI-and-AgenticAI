```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import ChatOpenAI
```

```python
llm_chain1 = ChatOpenAI(model = "gpt-4.1-nano", temperature=0.7)
llm_chain2 = ChatOpenAI(model = "o1-mini")

# Step-1: Get three obscure historical facts
initial_prompt = "Give me three obscure historical facts"
result_step_1 = llm_chain1.invoke(initial_prompt)
facts = result_step_1.content
print(f"Facts generated by LLM 1:\n{facts}\n")

# Step 2: Use the facts as input for the second LLM
final_prompt = f"Based on these facts:\n{facts}\nWhich one is the most surprising and why?"
result_step_2 = llm_chain2.invoke(final_prompt)
final_answer = result_step_2.content

print(f"Final answer from LLM 2:\n{final_answer}")
```
